```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn import metrics
```

```python
# making object of count Vectorizer
vectorizer = CountVectorizer(max_df=500, min_df=10,stop_words='english',max_features=50 )
```

```python
vectorizer
```

```python
# fiting sparse matrix and printing features name
X = vectorizer.fit_transform(data)
vectorizer.get_feature_names()
```

```python
print(X)
```

```python
X.toarray()
```

```python
import pandas as pd
```

```python
df= pd.DataFrame(X.toarray())
```

```python
df
```

```python
len(df.iloc[0])
```

```python
# i and j are indices of feature and storing their indices and score in list of lists [(i,j),MI score]
final=[]
place=[]
for i in range(len(df.iloc[0])):
    for j in range(i+1,len(df.iloc[0])):
        pos=(i,j)
#         print(df[i])
        score=metrics.mutual_info_score(df[i],df[j])

        place = [pos,score]
        final.append(place)
```

```python
final[0]
```

```python
import numpy as np
score_list=[]
for nm in final:
    score_list.append(nm[1])
score_list

        
```

```python
# making a descending sorting and get descended indices of score_list 
numbers = np. array(score_list)
n=len(score_list)
idx = np. argpartition(numbers, -n)[-n:]
indices = idx[np. argsort((-numbers)[idx])]
```

```python
indices
```

```python
# list with decraesing order of theri MI score
decreasing_MI_score=[]
for n in indices:
    print(final[n])
    decreasing_MI_score.append(final[n])
```

```python
len(decreasing_MI_score),len(final)
```

```python
decreasing_MI_score[0][0]
```

```python
# by using decreasing_MI_score[a][0][b]] list we extract words using theri indices which are in tuple form(i,j)
words_list=[]
tmp_list=[]
for a in range (len(decreasing_MI_score)):
    tmp_list=[]
    for b in range(2):
        tmp_list.append(vectorizer.get_feature_names()[decreasing_MI_score[a][0][b]])
    words_list.append(tmp_list)
#     print(vectorizer.get_feature_names()[decreasing_MI_score[a][0][1]],'\n')
```
