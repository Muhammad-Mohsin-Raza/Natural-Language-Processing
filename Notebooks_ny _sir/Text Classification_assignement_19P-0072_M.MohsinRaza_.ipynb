```python
data = open("Movies_TV.txt").read()
data = data.split('\n')
data.remove(data[0])
data.remove(data[-1])
data.remove(data[-1])
```

```python
data[0]
```

```python
reviews = []
y = []
for item in data:
    _,_, label, text = item.split('\t')
    reviews.append(text.strip())
    y.append(label.strip())
```

```python
reviews[0]
```

```python
y[0]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vec = TfidfVectorizer(max_df = 700, min_df = 10, ngram_range = (1,3), max_features = 8500,stop_words='english')
X = vec.fit_trahnsform(reviews)
X = X.toarray()
```

```python
len(vec.vocabulary_)
```

```python
(vec.get_feature_names())
```

```python
reviews[0]
```

```python
X[0]
```

```python
from sklearn.model_selection import train_test_split
trainX, testX, trainY, testY = train_test_split(X, y, shuffle = True, train_size=0.7)
```

```python
len(trainX), len(testX)
```

```python
from sklearn.linear_model import SGDClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
lc = SGDClassifier()
nbc = GaussianNB()
dtc = DecisionTreeClassifier()
knnc = KNeighborsClassifier()
```

```python
lc.fit(trainX, trainY)
nbc.fit(trainX, trainY)
dtc.fit(trainX, trainY)
knnc.fit(trainX, trainY)
```

```python
pred_y_lc = lc.predict(testX)
pred_y_nbc = nbc.predict(testX)
pred_y_dtc = dtc.predict(testX)
pred_y_knnc = knnc.predict(testX)
```

```python
from sklearn.metrics import accuracy_score
lc_acc = accuracy_score(testY, pred_y_lc)
nbc_acc = accuracy_score(testY, pred_y_nbc)
dtc_acc = accuracy_score(testY, pred_y_dtc)
knnc_acc = accuracy_score(testY, pred_y_knnc)
```

```python
print("Linear Classifier: ", lc_acc)
print("Naive Bayes Classifier: ", nbc_acc)
print("Decision Tree Classifier: ", dtc_acc)
print("KNN Classifier: ", knnc_acc)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
print(confusion_matrix(testY, pred_y_lc))
print(confusion_matrix(testY, pred_y_nbc))
print(confusion_matrix(testY, pred_y_dtc))
print(confusion_matrix(testY, pred_y_knnc))
```
